## EDA

```python
# Load data

import pandas as pd
import numpy as np
import os

df_analysisvariables = pd.read_csv(r'CSpine/CSV datasets/analysisvariables.csv')
df_clinicalpresentationsite = pd.read_csv(r'CSpine/CSV datasets/clinicalpresentationsite.csv')
df_injuryclassification = pd.read_csv(r'CSpine/CSV datasets/injuryclassification.csv')
df_demographics = pd.read_csv(r'CSpine/CSV datasets/demographics.csv')
df_analysisvariables.head()
```

```python
len(df_analysisvariables[df_analysisvariables['ControlType']=='case'])
```

```python
df_clinicalpresentationsite = df_clinicalpresentationsite[['CaseID','TotalGCS']]
df_clinicalpresentationsite.head()
```

```python
df_injuryclassification = df_injuryclassification[['CaseID','CSFractures']]
df_injuryclassification
```

```python
# Merge data

df = df_analysisvariables.merge(df_clinicalpresentationsite, on='CaseID', how='inner')
df = df.merge(df_injuryclassification, on='CaseID', how='inner')
df.head()
```

```python
df.to_csv('CSpine\CSV datasets\data.csv')
```

```python
df_demographics = df_demographics[['caseid','AgeInYears','Gender','Race','Ethnicity']]
df_demographics.head()
```

```python
df_demo = df.merge(df_demographics, left_on='CaseID', right_on='caseid', how='left')
df_demo.head()
```
